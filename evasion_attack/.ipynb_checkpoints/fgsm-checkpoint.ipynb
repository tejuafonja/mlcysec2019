```python
import os

import numpy as np
import matplotlib.pyplot as plt

import pickle
import time
from IPython.display import display, HTML

from mlcysec2019 import utils

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0)
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

%load_ext autoreload
%autoreload 2
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, TensorDataset
from torch.utils.data.sampler import SubsetRandomSampler

import torchvision.datasets as datasets
import torchvision.transforms as transforms
import torch.utils.data as data
import torchvision

import warnings 
warnings.filterwarnings("ignore")
```

```python
compute_mode = 'cpu'

if compute_mode == 'cpu':
    device = torch.device('cpu')
elif compute_mode == 'gpu':
    os.environ["CUDA_VISIBLE_DEVICES"] = '0'    
    device = torch.device('cuda')
else:
    raise ValueError('Unrecognized compute mode')
```

```python
transform = transforms.Compose(
    [transforms.Resize((32, 32)),
     transforms.ToTensor(),
     transforms.Normalize((0,), (1,))]
     )

train_data = datasets.MNIST('~/data', train=True, download=True, transform=transform)
test_data = datasets.MNIST('~/data', train=False, download=True, transform=transform)

train_ratio = 0.85
train_num = len(train_data)
randn_indices = torch.randperm(train_num)
train_val_split = int(train_num*train_ratio)

train_idxs, val_idxs = randn_indices[:train_val_split], randn_indices[train_val_split:]

train_sampler = SubsetRandomSampler(train_idxs)
val_sampler = SubsetRandomSampler(val_idxs)

train_loader = DataLoader(train_data, batch_size=32, shuffle=False, num_workers=2, sampler=train_sampler)
val_loader = DataLoader(train_data, batch_size=32, shuffle=False, num_workers=2, sampler=val_sampler)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False, num_workers=2)
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(1, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16*5*5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 16*5*5)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
pretrained_model = "../models/mnist_net_e6_a98.pkl"
```

```python
# Initialize the network
net = Net().to(device)

# Load the pretrained model
net.load_state_dict(torch.load(pretrained_model, map_location='cpu'))

# Set the model in evaluation mode. In this case this is for the Dropout layers
net.eval()


optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
criterion = nn.CrossEntropyLoss()
```

```python
# Generate images with noise
def fgsm_samples(imgs, epsilon, data_grad):
    adv_imgs = imgs + epsilon*data_grad.sign()
    adv_imgs = torch.clamp(adv_imgs, 0, 1)
    return adv_imgs

# FGSM attack
def fgsm_attack(net, criterion, imgs, labels, epsilon, adv_training=False):
    global device

    imgs.requires_grad = True

    net.eval()
    outputs = net(imgs)
    predicted_label = outputs.max(1, keepdim=False)[1]


    # Select correctly classified images to manupulate
    selected_labels = torch.where(
      predicted_label == labels,
      torch.tensor(1).to(device),
      torch.tensor(0).to(device)).byte()

    loss = criterion(outputs, labels)
    net.zero_grad()
    loss.backward()

    # Generate adversarial examples
    data_grad = imgs.grad[selected_labels].data
    adv_imgs = fgsm_samples(imgs[selected_labels], epsilon, data_grad)

    if adv_training:
        return adv_imgs, labels[selected_labels]

    else:
        outputs = net(adv_imgs)
        adv_labels = outputs.max(1, keepdim=False)[1]
        acc = float(torch.sum(adv_labels == labels[selected_labels]).cpu()) / adv_labels.size(0)

    # Select the successful adversarial examples
    selected_adv_labels = torch.where(
        adv_labels != labels[selected_labels],
        torch.tensor(1).to(device),
        torch.tensor(0).to(device)).byte()


    adversarials = []
    victim_imgs = imgs[selected_labels][selected_adv_labels]
    victim_labels = labels[selected_labels][selected_adv_labels]
    adv_imgs = adv_imgs[selected_adv_labels]
    adv_labels = adv_labels[selected_adv_labels]

    for victim_labels, victim_imgs, adv_labels, adv_imgs in zip(
        victim_labels, victim_imgs, adv_labels, adv_imgs):

        victim_labels = int(victim_labels.cpu().detach().numpy())
        adv_labels = int(adv_labels.cpu().detach().numpy())
        victim_imgs = victim_imgs.cpu().detach().numpy()
        adv_imgs = adv_imgs.cpu().detach().numpy()
        adversarials.append((victim_labels, victim_imgs, adv_labels, adv_imgs))

    return acc, adversarials


def fgsm_batch_attack(net, criterion, data_loader, epsilon):


    global device

    adversarials = []

    total_acc = 0
    batch_size = data_loader.batch_size

    for imgs, labels in data_loader:
        imgs, labels = imgs.to(device), labels.to(device)

        acc, advs = fgsm_attack(net, criterion, imgs, labels, epsilon)
        adversarials.extend(advs)
        total_acc += acc



    total_acc /= float(len(data_loader))

    return adversarials, total_acc
```

```python
epsilons = np.linspace(0.05, 0.5, 10)
fgsm_advs = []
fgsm_acc = []

for eps in epsilons:

    start = time.time()
    _adv, _acc = fgsm_batch_attack(net, criterion, test_loader, eps)
    end = time.time()

    print('eps: {:.2f}, num of samples: {}, acc: {:.3f}, time spent:{:.3f} secs'.format(
      eps, len(_adv), _acc, end-start))

    fgsm_advs.append(_adv)
    fgsm_acc.append(_acc)
```

```python
plt.plot(epsilons, fgsm_acc)
plt.scatter(epsilons, fgsm_acc)
plt.grid()
plt.xlabel('epsilon')
plt.ylabel('accuracy')
plt.title("Epsilon Vs Accuracy")
```

```python
# Adversarial sample set
for i, eps in enumerate(epsilons):
    print("FGSM with epsilon {:.3f} has generated {} samples".format(eps, len(fgsm_advs[i])))
```

The lower the epsilon value, the harder it is to fool the victim's model
